Ahora, acerca la imagen a un punto de lanzamiento y explora sus alrededores para ver si puedes encontrar fácilmente alguna vía férrea, carretera, línea costera, etc. Mueve el ratón a estos puntos y anota sus coordenadas (que se muestran en la parte superior izquierda) para calcular la distancia al punto de lanzamiento.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
