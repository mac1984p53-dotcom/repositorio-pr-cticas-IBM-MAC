## Tareas

Ahora escribe y ejecuta consultas SQL para resolver las tareas asignadas.

### Tarea 1

##### Muestra los nombres de los sitios de lanzamiento únicos en la misión espacial.


```python
res = cur.execute("SELECT DISTINCT Launch_Site FROM SPACEXTBL")  
tabla = prettytable.from_db_cursor(res)
print(tabla)
```


### Task 2


##### Mostrar 5 registros donde los sitios de lanzamiento comiencen con la cadena 'CCA'


```python
# 1. Ejecutamos la consulta con el filtro LIKE y el límite
res = cur.execute("SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5")

# 2. Mostramos el resultado con PrettyTable
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 3




##### Muestra la masa total de carga útil transportada por los cohetes lanzadores de la NASA (CRS)


```python
# Ejecutar la consulta filtrando por NASA (CRS)
res = cur.execute("SELECT SUM(PAYLOAD_MASS__KG_) AS Masa_Total_CRS FROM SPACEXTBL WHERE Customer = 'NASA (CRS)'")

# Mostrar con PrettyTable
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 4




##### Mostrar la masa promedio de carga útil transportada por la versión F9 v1.1


```python
# Ejecutar la consulta para calcular el promedio
res = cur.execute("SELECT AVG(PAYLOAD_MASS__KG_) AS Masa_Promedio FROM SPACEXTBL WHERE Booster_Version = 'F9 v1.1'")

# Mostrar con PrettyTable
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 5

##### Indica la fecha en que se logró el primer aterrizaje exitoso en la plataforma de aterrizaje.

Sugerencia: Usa la función min.


```python
# Ejecutar la consulta con MIN para obtener la primera fecha
res = cur.execute("SELECT MIN(Date) AS Primer_Aterrizaje_Exitoso FROM SPACEXTBL WHERE Landing_Outcome = 'Success (ground pad)'")

# Mostrar con PrettyTable
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 6

##### Enumere los nombres de los propulsores que han tenido éxito en barcos no tripulados y tienen una masa de carga útil mayor de 4000 pero menor de 6000.


```python
# Ejecutar la consulta con los tres filtros solicitados
query = """
SELECT Booster_Version 
FROM SPACEXTBL 
WHERE Landing_Outcome = 'Success (drone ship)' 
AND PAYLOAD_MASS__KG_ > 4000 
AND PAYLOAD_MASS__KG_ < 6000;
"""

res = cur.execute(query)

# Mostrar el resultado
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 7




##### Enumere el número total de resultados de misiones exitosas y fallidas


```python
# Ejecutamos la consulta agrupando por el resultado de la misión
res = cur.execute("SELECT Mission_Outcome, COUNT(*) AS Total FROM SPACEXTBL GROUP BY Mission_Outcome")

# Mostrar con PrettyTable
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 8



##### Enumera todas las versiones de booster que han transportado la masa máxima de carga útil, utilizando una subconsulta con una función de agregación adecuada.


```python
# Definimos la consulta con la subconsulta integrada
query = """
SELECT DISTINCT Booster_Version, PAYLOAD_MASS__KG_
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL);
"""

res = cur.execute(query)

# Mostramos el resultado
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 9


##### Lista los registros que mostrarán los nombres de los meses, los resultados de aterrizaje fallidos en el barco no tripulado, las versiones de los propulsores y el sitio de lanzamiento para los meses del año 2015.

**Nota: SQLite no admite nombres de meses. Por lo tanto, debe usar `substr(Fecha, 6, 2)` para obtener los meses y `substr(Fecha, 0, 5) = '2015'` para el año.**


```python
# Consulta SQL siguiendo las indicaciones de substr
query = """
SELECT 
    substr(Date, 6, 2) AS Mes, 
    Landing_Outcome, 
    Booster_Version, 
    Launch_Site 
FROM SPACEXTBL 
WHERE Landing_Outcome = 'Failure (drone ship)' 
AND substr(Date, 1, 4) = '2015';
"""

res = cur.execute(query)

# Mostrar resultados formateados
tabla = prettytable.from_db_cursor(res)
print(tabla)
```

### Task 10




##### Clasifique el número de resultados de aterrizaje (como Fallo (nave no tripulada) o Éxito (plataforma terrestre)) entre las fechas 2010-06-04 y 2017-03-20, en orden descendente.


```python
query = """
SELECT Landing_Outcome, COUNT(*) AS Conteo 
FROM SPACEXTBL 
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' 
GROUP BY Landing_Outcome 
ORDER BY Conteo DESC;
"""

res = cur.execute(query)
tabla = prettytable.from_db_cursor(res)
print(tabla)
```
